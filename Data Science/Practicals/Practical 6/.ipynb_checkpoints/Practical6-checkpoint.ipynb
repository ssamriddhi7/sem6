```python
# Practical 6 - Data Handling

import pandas as pd

df = pd.read_csv('workout.csv', na_values=['None', ''])
df['Age'].fillna(df['Age'].median())

workout_typo = { 'weightss': 'weights' }
df['Type_of_Workout'] = df['Type_of_Workout'].str.lower()
df['Type_of_Workout'] = df['Type_of_Workout'].replace(workout_typo)
df['Type_of_Workout'] = df['Type_of_Workout'].str.capitalize()

df['Occupation'].fillna('Star Parivar')

df['Gender'].fillna('Other')
df['Gender'] = df['Gender'].str.title()

print(df)
```

```python
import pandas as pd
from datetime import datetime

def update_salary(df):
    df['IT'] = df['salary'] * 0.10
    df['TA'] = df['salary'] * 0.05
    df['DA'] = df['salary'] * 0.03
    df['HRA'] = df['salary'] * 0.07
    df['gross_salary'] = df['salary'] + df['IT'] + df['TA'] + df['DA'] + df['HRA']

df = pd.read_csv('employee.csv', na_values=['None', ''])

df['name'].fillna('Mithibai')

df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'].fillna(df['age'].median())

df['gender'].fillna('Others')

df['department'].fillna('CS')

df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df['salary'].fillna(df['salary'].median())

df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
df['join_date'] = df['join_date'].dt.strftime('%Y-%m-%d')

df['status'].fillna('probation')

df['email'].fillna('star@mithibai.com')

print(f"Unique Employee IDs: {df['employee_id'].nunique()}")

update_salary(df)

current_date = pd.to_datetime('2025-01-30')

df['join_date'] = pd.to_datetime(df['join_date'], format='%Y-%m-%d', errors='coerce')

df['years_of_experience'] = (current_date - df['join_date']).dt.days / 365
df.loc[df['years_of_experience'] > 10, 'salary'] = df['salary'] * 1.12
update_salary(df)
df.drop(['IT', 'TA', 'DA', 'HRA'], axis=1)

print(df)
```
