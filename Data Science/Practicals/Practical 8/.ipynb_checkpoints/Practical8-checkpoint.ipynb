```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value': [10, 25, 35, 20, 10]}
df = pd.DataFrame(data)
plt.figure(figsize=(7, 7))
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

days = np.array([1, 2, 3, 4, 5, 6, 7])
temperature = np.array([22, 24, 21, 23, 25, 28, 30])
df_line = pd.DataFrame({
    'Day': days,
    'Temperature': temperature
})
plt.figure(figsize=(10, 6))
plt.plot(df_line['Day'], df_line['Temperature'], marker='o', linestyle='-', color='b', label='Temperature')
plt.title('Temperature Over the Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```
