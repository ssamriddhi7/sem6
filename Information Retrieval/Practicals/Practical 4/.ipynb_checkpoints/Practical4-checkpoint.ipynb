```python
# Practical 4 - Inverted Index

from collections import defaultdict

def inverted_index():
    index = defaultdict(list)
    for i in range(1, 6):
        with open(f"../Practical 1 and 2/filtered{i}.txt", 'r') as f: text = set(f.read().split())
        for word in text: index[word].append(i)
    return dict(sorted(index.items()))

index = inverted_index()
for key in index:
    print(f"{key}: {index[key]}")
```
